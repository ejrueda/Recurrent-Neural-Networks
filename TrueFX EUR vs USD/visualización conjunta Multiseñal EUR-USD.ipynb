```python
import numpy as np
import pandas as pd
import bokeh.plotting as bp
from bokeh.layouts import gridplot
#bp.output_notebook()
```

se cargan los resultados obtenidos por la multiseñal EUR/USD,GBP/USD -> EUR/USD

```python
result_KNC = pd.read_csv('../resultados/TrueFX/EUR-USD/MultiSeñal/KNeighborsClassifier/pnl-EUR-USD_GBP-USD_KNC_wsize3_2018-01.csv',
                        parse_dates=True, infer_datetime_format=True, index_col=0)
result_DTC = pd.read_csv('../resultados/TrueFX/EUR-USD/MultiSeñal/DecisionTreeClassifier/pnl-EUR-USD_GBP-USD_DTC_wsize3_2018-01.csv',
                        parse_dates=True, infer_datetime_format=True, index_col=0)
result_RFC = pd.read_csv('../resultados/TrueFX/EUR-USD/MultiSeñal/RandomForestClassifier/pnl-EUR-USD_GBP-USD_RFC_wsize3_2018-01.csv',
                        parse_dates=True, infer_datetime_format=True, index_col=0)
```

```python
result_DTC.head()
```

```python
bp.reset_output()
```

```python
plot_1 = bp.figure(width=850, height=300,title='Profit and Loss',x_axis_type='datetime')
#es el mismo indice para todos entonces no hay problema
x = result_RFC.index
plot_1.line(x, result_RFC.iloc[:,0], color='blue', legend='RFC', alpha=0.5)
plot_1.circle(x, result_RFC.iloc[:,0], color='blue', alpha=0.5,size=3)
plot_1.line(x, result_DTC.iloc[:,0], color='red', legend='DTC', alpha=0.5)
plot_1.circle(x, result_DTC.iloc[:,0], color='red',alpha=0.5,size=3)
plot_1.line(x, result_KNC.iloc[:,0], color='green', legend='KNC', alpha=0.5)
plot_1.circle(x, result_KNC.iloc[:,0], color='green',alpha=0.5,size=3)
plot_1.line(x, np.zeros(len(result_RFC)), color='black', legend='PNL = 0', line_dash='dotted')
plot_1.xaxis.axis_label = 'datetime'
plot_1.yaxis.axis_label = 'PNL'
plot_1.legend.location ='bottom_center'
plot_1.legend.orientation = 'horizontal'

plot_2 = bp.figure(width=550, height=250, title='buys per day', x_axis_type='datetime')
plot_2.line(x, result_RFC.iloc[:,1], color='blue')
plot_2.line(x, result_DTC.iloc[:,1], color='red')
plot_2.line(x, result_KNC.iloc[:,1], color='green')
#p2_RFC.circle(x, result_RFC.iloc[:,1], color='green')
plot_2.xaxis.axis_label = 'datetime'
plot_2.yaxis.axis_label = 'buys per day'

#se utiliza el x_range para correlacionar las dos gráficas por el eje x
plot_3 = bp.figure(width=550, height=250, title='sells per day', x_axis_type='datetime',x_range=plot_2.x_range)
plot_3.line(x, result_RFC.iloc[:,2], color='blue')
plot_3.line(x, result_DTC.iloc[:,2], color='red')
plot_3.line(x, result_KNC.iloc[:,2], color='green')
#p3_RFC.circle(x, result_RFC.iloc[:,2], color='red')
plot_3.xaxis.axis_label = 'datetime'
plot_3.yaxis.axis_label = 'sells per day'

#para pintar el PNL que se va acumulando del KNC
PNL_accu_KNC = []
count_KNC = 0
for i in result_KNC.iloc[:,0]:
    count_KNC += i
    PNL_accu_KNC.append(count_KNC)
#para pintar el PNL que se va acumulando del DTC
PNL_accu_DTC = []
count_DTC = 0
for i in result_DTC.iloc[:,0]:
    count_DTC += i
    PNL_accu_DTC.append(count_DTC) 
#para pintar el PNL que se va acumulando del RFC
PNL_accu_RFC = []
count_RFC = 0
for i in result_RFC.iloc[:,0]:
    count_RFC += i
    PNL_accu_RFC.append(count_RFC)
    
plot_4 = bp.figure(width=850, height=300,title='PNL accumulated',x_axis_type='datetime',x_range=plot_1.x_range)
plot_4.line(x, PNL_accu_RFC,color='blue',legend='RFC',alpha=0.5)
plot_4.circle(x, PNL_accu_RFC, color='blue', alpha=0.5, size=3)
plot_4.line(x, PNL_accu_DTC,color='red',legend='DTC',alpha=0.5)
plot_4.circle(x, PNL_accu_DTC, color='red', alpha=0.5, size=3)
plot_4.line(x, PNL_accu_KNC,color='green',legend='KNC',alpha=0.5)
plot_4.circle(x, PNL_accu_KNC, color='green', alpha=0.5, size=3)
plot_4.line(x, np.zeros(len(result_RFC)), color='black', legend='PNL = 0', line_dash='dotted')
plot_4.xaxis.axis_label = 'datetime'
plot_4.yaxis.axis_label = 'PNL accumulated'
plot_4.legend.location ='bottom_center'
plot_4.legend.orientation = 'horizontal'

bp.show(gridplot([[plot_1],[plot_4],[plot_2,plot_3]], toolbar_location="right"))
bp.output_file(filename='../resultados/TrueFX/EUR-USD/MultiSeñal/comparaciónMultiSeñal'
               , title='comparación Multiseñal')
```

```python
result_KNC_MS = pd.read_csv('../resultados/TrueFX/EUR-USD/KNeighborsClassfier/pnl_KNC_wsize3_2018-01.csv',
                           parse_dates=True, infer_datetime_format=True, index_col=0)
result_DTC_MS = pd.read_csv('../resultados/TrueFX/EUR-USD/DecisionTreeClassifier/pnl_DTC_wsize3_2018-01.csv',
                           parse_dates=True, infer_datetime_format=True, index_col=0)
result_RFC_MS = pd.read_csv('../resultados/TrueFX/EUR-USD/RandomForestClassifier/pnl_RFC_wsize3_2018-01.csv',
                           parse_dates=True, infer_datetime_format=True, index_col=0)
```

donde la linea punteada es la señal original EUR/USD y la linea normal es la multiseñal EUR/USD,GBP/USD..

```python
plot_1 = bp.figure(width=850, height=350,title='Profit and Loss EUR/USD & EUR-GBP/USD',x_axis_type='datetime')
#es el mismo indice para todos entonces no hay problema
x = result_RFC.index
#los datos de la señal sola EUR-USD
plot_1.line(x, result_RFC.iloc[:,0], color='#0EE7E7', legend='RFC',line_dash='dotted')
plot_1.circle(x, result_RFC.iloc[:,0], color='#0EE7E7',size=3)
plot_1.line(x, result_DTC.iloc[:,0], color='red', legend='DTC',line_dash='dotted')
plot_1.circle(x, result_DTC.iloc[:,0], color='red',size=3)
plot_1.line(x, result_KNC.iloc[:,0], color='green', legend='KNC',line_dash='dotted')
plot_1.circle(x, result_KNC.iloc[:,0], color='green',size=3)
#datos multiseñal EUR-USD, GBP-USD
plot_1.line(x, result_RFC_MS.iloc[:,0], color='#0EE7E7', legend='RFC', alpha=1)
plot_1.circle(x, result_RFC_MS.iloc[:,0], color='#0EE7E7', alpha=1,size=3)
plot_1.line(x, result_DTC_MS.iloc[:,0], color='red', legend='DTC', alpha=1)
plot_1.circle(x, result_DTC_MS.iloc[:,0], color='red',alpha=1,size=3)
plot_1.line(x, result_KNC_MS.iloc[:,0], color='green', legend='KNC', alpha=1)
plot_1.circle(x, result_KNC_MS.iloc[:,0], color='green',alpha=1,size=3)

plot_1.line(x, np.zeros(len(result_RFC)), color='black', legend='PNL = 0')
plot_1.xaxis.axis_label = 'datetime'
plot_1.yaxis.axis_label = 'PNL'
plot_1.legend.location ='bottom_center'
plot_1.legend.orientation = 'horizontal'

plot_2 = bp.figure(width=550, height=250, title='buys per day EUR/USD & EUR-GBP/USD', x_axis_type='datetime')
#compras de la señal EUR/USD
plot_2.line(x, result_RFC.iloc[:,1], color='#0EE7E7', line_dash='dotted')
plot_2.line(x, result_DTC.iloc[:,1], color='red', line_dash='dotted')
plot_2.line(x, result_KNC.iloc[:,1], color='green', line_dash='dotted')
#compras de la multiseñal EUR/USD,GBP/USD
plot_2.line(x, result_RFC_MS.iloc[:,1], color='#0EE7E7')
plot_2.line(x, result_DTC_MS.iloc[:,1], color='red')
plot_2.line(x, result_KNC_MS.iloc[:,1], color='green')
plot_2.xaxis.axis_label = 'datetime'
plot_2.yaxis.axis_label = 'buys per day'

#se utiliza el x_range para correlacionar las dos gráficas por el eje x
plot_3 = bp.figure(width=550, height=250, title='sells per day EUR/USD & EUR-GBP/USD', x_axis_type='datetime',x_range=plot_2.x_range)
#ventas de la señal EUR/USD
plot_3.line(x, result_RFC.iloc[:,2], color='blue', line_dash='dotted')
plot_3.line(x, result_DTC.iloc[:,2], color='red', line_dash='dotted')
plot_3.line(x, result_KNC.iloc[:,2], color='green', line_dash='dotted')
#ventas de la multiseñal EUR/USD,GBP/USD
plot_3.line(x, result_RFC_MS.iloc[:,2], color='#0EE7E7')
plot_3.line(x, result_DTC_MS.iloc[:,2], color='red')
plot_3.line(x, result_KNC_MS.iloc[:,2], color='green')
plot_3.xaxis.axis_label = 'datetime'
plot_3.yaxis.axis_label = 'sells per day'

#para pintar el PNL que se va acumulando del KNC
PNL_accu_KNC = []
count_KNC = 0
for i in result_KNC.iloc[:,0]:
    count_KNC += i
    PNL_accu_KNC.append(count_KNC)
#para pintar el PNL que se va acumulando del DTC
PNL_accu_DTC = []
count_DTC = 0
for i in result_DTC.iloc[:,0]:
    count_DTC += i
    PNL_accu_DTC.append(count_DTC) 
#para pintar el PNL que se va acumulando del RFC
PNL_accu_RFC = []
count_RFC = 0
for i in result_RFC.iloc[:,0]:
    count_RFC += i
    PNL_accu_RFC.append(count_RFC)

#para pintar el PNL que se va acumulando del KNC MULTISEÑAL
PNL_accu_KNC_MS = []
count_KNC_MS = 0
for i in result_KNC_MS.iloc[:,0]:
    count_KNC_MS += i
    PNL_accu_KNC_MS.append(count_KNC_MS)
#para pintar el PNL que se va acumulando del DTC MULTISEÑAL
PNL_accu_DTC_MS = []
count_DTC_MS = 0
for i in result_DTC_MS.iloc[:,0]:
    count_DTC_MS += i
    PNL_accu_DTC_MS.append(count_DTC_MS) 
#para pintar el PNL que se va acumulando del RFC MULTISEÑAL
PNL_accu_RFC_MS = []
count_RFC_MS = 0
for i in result_RFC_MS.iloc[:,0]:
    count_RFC_MS += i
    PNL_accu_RFC_MS.append(count_RFC_MS)
    
plot_4 = bp.figure(width=850, height=350,title='PNL accumulated EUR/USD & EUR-GBP/USD',x_axis_type='datetime',x_range=plot_1.x_range)
#para pintar el de la señal EUR/USD
plot_4.line(x, PNL_accu_RFC,color='#0EE7E7',legend='RFC',line_dash='dotted')
plot_4.circle(x, PNL_accu_RFC, color='#0EE7E7',size=3)
plot_4.line(x, PNL_accu_DTC,color='red',legend='DTC',line_dash='dotted')
plot_4.circle(x, PNL_accu_DTC, color='red',size=3)
plot_4.line(x, PNL_accu_KNC,color='green',legend='KNC',line_dash='dotted')
plot_4.circle(x, PNL_accu_KNC, color='green',size=3)
#para pintar el de la MULTISEÑAL EUR/USD,GBP/USD
plot_4.line(x, PNL_accu_RFC_MS,color='#0EE7E7',legend='RFC',alpha=1)
plot_4.circle(x, PNL_accu_RFC_MS, color='#0EE7E7', alpha=1, size=3)
plot_4.line(x, PNL_accu_DTC_MS,color='red',legend='DTC',alpha=1)
plot_4.circle(x, PNL_accu_DTC_MS, color='red', alpha=1, size=3)
plot_4.line(x, PNL_accu_KNC_MS,color='green',legend='KNC',alpha=1)
plot_4.circle(x, PNL_accu_KNC_MS, color='green', alpha=1, size=3)

plot_4.line(x, np.zeros(len(result_RFC)), color='black', legend='PNL = 0')
plot_4.xaxis.axis_label = 'datetime'
plot_4.yaxis.axis_label = 'PNL accumulated'
plot_4.legend.location ='bottom_center'
plot_4.legend.orientation = 'horizontal'

bp.show(gridplot([[plot_1],[plot_4],[plot_2,plot_3]], toolbar_location="right"))
bp.output_file(filename='../resultados/TrueFX/EUR-USD/MultiSeñal/ComparacionConjuntaMSconSeñalSimple'
               , title='comparaciónConjunta Multiseñal')
```
