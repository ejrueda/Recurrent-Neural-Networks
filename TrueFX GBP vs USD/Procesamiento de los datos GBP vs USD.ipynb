```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#leer los datos obtenidos de TrueFX
datos = pd.read_csv("../data/TrueFX/GBP-USD/GBPUSD-2018-01.csv",float_precision='high',
                    names =list(['date','bid', 'ask']))
print(datos.shape)
datos.head()
```

```python
plt.figure(figsize=(15,5))
plt.title('spread')
plt.plot(range(datos.shape[0]),datos.iloc[:,2]-datos.iloc[:,1], color='green');
```

```python
plt.figure(figsize=(15,5))
plt.title('Bid')
plt.plot(range(datos.shape[0]),datos.iloc[:,1]);
```

```python
datos = datos.set_index(pd.to_datetime(datos.iloc[:,0])) #Se indexan los datos por su fecha y hora
datos = datos.drop(['date'],axis = 1) #se elimina la columna date
datos.head()
```

```python
#se agrupan los indices segn su fecha, como hay fechas repetidas de toma la que primero est
datos = datos.groupby(datos.index).first()
datos.shape
```

```python
np.unique(datos.index.date)
```

```python
datos['2018-01-07']
```

```python
print('tiempo que los datos estn corridos:')
pd.Timedelta(hours=23,minutes=59,seconds=42) - pd.Timedelta(hours=22,minutes=2,seconds=46)
```

```python
#para correr los dias, quedan 5 bussines day
datos.index = datos.index + pd.Timedelta(hours=1, minutes=59, seconds=24)
```

```python
np.unique(datos.index.date)
```

```python
#se muestrean en intervalos de 10s
datos = datos.resample('5T', loffset='5T').last() #se coge el ltimo valor de muestreo de los 10s
print(datos.shape)
datos.head()
```

```python
plt.figure(figsize=(15,5))
plt.title('Bid')
plt.xlabel('time 10s')
plt.ylabel('GBP/USD')
plt.plot(np.arange(datos.shape[0]), datos.iloc[:,0],color='blue',alpha=0.6, label='Bid')
plt.legend();
```

```python
print('datos nan')
print('------------')
print(np.sum(pd.isnull(datos)*1))
```

```python
datos = datos.dropna() #se eliminan los nan
print(datos.shape)
```

```python
plt.figure(figsize=(15,5))
plt.title('Bid')
plt.xlabel('time 10s')
plt.ylabel('GBP/USD')
plt.plot(np.arange(datos.shape[0]), datos.iloc[:,0],color='blue',alpha=0.6, label='Bid')
plt.legend();
```

```python
datos.to_csv('../data/TrueFX/GBP-USD/datos_procesados_5T_GBPUSD-2018-01.csv')
```
