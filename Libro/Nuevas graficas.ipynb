```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

para graficar los pnl

```python
x = [10,20, 30, 40, 50, 60, 90, 300]
gnb = [-0.27, -0.1396, -0.0952, -0.0719, -0.0596, -0.0501, -0.0332, -0.0054]
kn = [-0.2724, -0.1415, -0.0982, -0.0772, -0.0637, -0.0547, -0.0363, -0.0106]
dt = [-0.2710, -0.1443, -0.0989, -0.0775, -0.0628, -0.0492, -0.037, -0.0104]
rf = [-0.2709, -0.1401, -0.1012, -0.0761, -0.0612, -0.0513, -0.0366, -0.0083]
```

```python
plt.figure(figsize=(5,8))
ax = plt.axes()
plt.plot(x,gnb, label='GNB')
plt.plot(x,kn, label='KNC')
plt.plot(x,dt, label='DTC')
plt.plot(x,rf, label='RFC')
plt.grid()
plt.legend(loc='center right')
ax.spines['right'].set_visible(False)
plt.title('EUR/USD')
plt.ylabel('PNL')
plt.xlabel('Segundos')
ax.spines['top'].set_visible(False)
```
