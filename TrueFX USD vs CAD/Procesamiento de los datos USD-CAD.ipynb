```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#leer los datos obtenidos de TrueFX#leer l 
datos = pd.read_csv("../data/TrueFX/USD-CAD/USDCAD-2018-01.csv",float_precision='high',parse_dates=True
                    ,names =list(['ref','bid', 'ask']),infer_datetime_format=True,index_col=1)
datos.shape
```

```python
datos = datos.drop('ref',axis=1)
datos.head()
```

```python
#se agrupan los indices según su fecha, como hay fechas repetidas de toma la que primero está
datos = datos.groupby(datos.index).first()
datos.shape
```

```python
np.unique(datos.index.date)
```

```python
datos['2018-1-7']
```

```python
pd.Timedelta(hours=23,minutes=59,seconds=51) - pd.Timedelta(hours=22,minutes=2,seconds=30)
```

```python
#para correr los dias, quedan 5 bussines day
datos.index = datos.index + pd.Timedelta(hours=1, minutes=57, seconds=30)
```

```python
np.unique(datos.index.date)
```

```python
#se muestrean en intervalos de 10s 
data = datos.resample('5T', loffset='5T').last().dropna() #se coge el último valor de muestreo de los 10s
print(data.shape)
data.head()
```

```python
#guardar el data set pre-procesado
data.to_csv('../data/TrueFX/USD-CAD/datos_procesados_5T_USDCAD-2018-01.csv')
```
