```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import bokeh.plotting as bp
%matplotlib inline
```

```python
datos_1 = pd.read_csv('./data/TrueFX/EUR-USD/datos_procesados_EURUSD-2018-01.csv', index_col=0
                    ,infer_datetime_format=True, parse_dates=True)
datos_2 = pd.read_csv('./data/TrueFX/GBP-USD/GBPUSD-preprocesado-2018-01.csv', index_col=0
                    ,infer_datetime_format=True, parse_dates=True)
```

```python
y1 = datos_1.iloc[:20,0]
y2 = datos_2.iloc[:20,0] - 0.1457
```

```python
plt.figure(figsize=(15,5))
plt.title('Comportamiento del Bid', size=15)
plt.ylabel('EUR/USD', size=12)
plt.plot(range(len(y1)),y1,label='EUR/USD')
plt.plot(range(len(y2)),y2,label='GBP/USD')
plt.xticks(range(len(y1)), y1.index.time,size=10,rotation=70)
#plt.plot(range(len(y)),datos.iloc[6:11,1], label='Ask',linewidth=5, color='purple')
#plt.yscale('log')
plt.grid()
plt.legend();
```

```python
p1= bp.Figure(width=1300, height=600,title='bid & ask',x_axis_type='datetime',
               title_location = 'above')
p1.line(datos_1.index, datos_1.iloc[:,0], color='green', legend='bid')
p1.line(datos_1.index, datos_1.iloc[:,1], color='blue', legend='ask')
p1.xaxis.axis_label = 'Datetime'
p1.yaxis.axis_label = 'EUR/USD'
bp.show(p1)
```
