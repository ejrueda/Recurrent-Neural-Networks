```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#leer los datos obtenidos de TrueFX
datos = pd.read_csv("./data/TrueFX/EUR-USD/EURUSD-2017-09.csv",float_precision='high', names =list(['date','bid', 'ask']))
datos.shape
```

```python
plt.figure(figsize=(15,5))
plt.plot(range(datos.shape[0]),datos.iloc[:,1])
```

```python
datos.head()
```

```python
datos = datos.set_index(pd.to_datetime(datos.iloc[:,0])) #Se indexan los datos por su fecha y hora
datos = datos.drop(['date'],axis = 1) #se elimina la columna date
datos.head()
```

```python
#se agrupan los indices según su fecha, como hay fechas repetidas de toma la que primero está
datos = datos.groupby(datos.index).first()
datos.shape
```

```python
np.unique(datos.index.date)
```

```python
datos['2017-9-17']
```

```python
pd.Timedelta(hours=24,minutes=0,seconds=0) - pd.Timedelta(hours=21,minutes=0,seconds=39)
```

```python
#para correr los dias, quedan 5 bussines day
datos.index = datos.index + pd.Timedelta(hours=2, minutes=59, seconds=21)
```

```python
datos.head()
```

```python
np.unique(datos.index.date)
```

```python
datos = datos.drop(datos['2017-9-17'].index) #se elimina este dato que quedó volando
```

```python
#se muestrean en intervalos de 10s
datos = datos.resample('10s', loffset='10s').last() #se coge el último valor de muestreo de los 10s
print(datos.shape)
datos.head()
```

```python
plt.figure(figsize=(15,5))
plt.title('Bid')
plt.xlabel('time 10s')
plt.ylabel('Bid')
plt.plot(np.arange(datos.shape[0]), datos.iloc[:,0],color='blue',alpha=0.6, label='Bid')
plt.legend();
```

```python
print('datos nan')
print('------------')
print(np.sum(pd.isnull(datos)*1))
```

```python
datos = datos.dropna() #se eliminan los nan
print(datos.shape)
```

```python
#guardar el data set pre-procesado
datos.to_csv('data/TrueFX/EUR-USD/datos_procesados_EURUSD-2017-09.csv')
```
